```python
#most predictable pattern of temp was between -20 and 20 lat the temperature was the highest

#there are to little other factors to be able to get good analysis of other variables we pulled in,
#windspeed and clouds would be more used along with longitude to be able to see how they are affected over land masses






# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
# Import API key
from api_keys import weatherkey

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weatherkey

City = []
Clouds = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

print(f"Beginning Data Retrieval")
print(f"__________________________")

for city in cities:
    
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        City.append(response["name"])
        Clouds.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        
        record= record + 1
        
        
        time.sleep(1.5)
        
   
    except:
        print("City not found. Skipping...")
    continue
    
print(f"Data Retrieval Ended")    
  
```

```python
weather_dict = {"City": City,"Cloudiness":Clouds,"Country":country,"Date":date,"Humidity": humidity,
    "Lat":lat,"Lng":lng,"Max Temp": max_temp,"Wind Speed":wind_speed}


city_weather = pd.DataFrame(weather_dict)

 
city_weather.count()
```

```python

city_weather.to_csv("city_weather.csv")

city_weather.head()
```

```python
#lat v Temp
plt.scatter(city_weather['Lat'],city_weather['Max Temp'],marker='o')
plt.title('Lat VS Max Temp')
plt.ylabel('Temperature (F)')
plt.xlabel('Lat')

plt.savefig('Lat_VS_MaxTemp.png')
plt.show()
```

```python
#lat V Clouds
plt.scatter(city_weather['Lat'],city_weather['Cloudiness'],marker='o')
plt.title('Lat VS Cloudiness')
plt.ylabel('Cloudiness %')
plt.xlabel('Lat')

plt.savefig('Lat_VS_Clouds.png')
plt.show()
```
